```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
def MostShortedStocks():
    # Get Urls
    results = [i for i in range(0,4500,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/most_shorted_stocks?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/most_shorted_stocks?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def UndervaluedGrowthStoks():
    # Get Urls
    results = [i for i in range(0,300,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/undervalued_growth_stocks?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/undervalued_growth_stocks?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def GrowthTechnologyStoks():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    # base_url = 'https://finance.yahoo.com/screener/predefined/growth_technology_stocks?count=100&offset=0'
    url = 'https://finance.yahoo.com/screener/predefined/growth_technology_stocks?offset=0&count={}'

    url_list = []
    # url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def DayGainers():
    # Get Urls
    results = [i for i in range(0,300,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/day_gainers?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/day_gainers?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def DailyLosers():
    # Get Urls
    results = [i for i in range(0,300,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/day_losers?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/day_losers?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def MostActives():
    # Get Urls
    results = [i for i in range(0,300,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/most_actives?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/most_actives?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def UndervaluedLargeCaps():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    # base_url = 'https://finance.yahoo.com/screener/predefined/undervalued_large_caps?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/undervalued_large_caps?offset=0&count={}'

    url_list = []
    # url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def AggressiveSmallCaps():
    # Get Urls
    results = [i for i in range(0,500,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/aggressive_small_caps?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/aggressive_small_caps?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def SmallCapGainers():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    # base_url = 'https://finance.yahoo.com/screener/predefined/small_cap_gainers?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/small_cap_gainers?offset=0&count={}'

    url_list = []
    # url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def TopMutualFunds():
    # Get Urls
    results = [i for i in range(0,1700,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/top_mutual_funds?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/top_mutual_funds?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def PortfolioAnchors():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/portfolio_anchors?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/portfolio_anchors?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def SolidLargeGrowthFunds():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/solid_large_growth_funds?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/solid_large_growth_funds?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def SolidMidCapGrowthFunds():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    # base_url = 'https://finance.yahoo.com/screener/predefined/solid_midcap_growth_funds?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/solid_midcap_growth_funds?offset=0&count={}'

    url_list = []
    # url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def ConservativeForeignFunds():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    base_url = 'https://finance.yahoo.com/screener/predefined/conservative_foreign_funds?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/conservative_foreign_funds?count=100&offset={}'

    url_list = []
    url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df


def HighYieldBond():
    # Get Urls
    results = [i for i in range(0,200,100)][1:]

    # base_url = 'https://finance.yahoo.com/screener/predefined/high_yield_bond?offset=0&count=100'
    url = 'https://finance.yahoo.com/screener/predefined/high_yield_bond?offset=0&count={}'

    url_list = []
    # url_list.append(base_url)
    for i in results:
        url_list.append(url.format(i))

    # Get html tables
    df_list = []
    for i in url_list:
        table = pd.read_html(i)[0]
        df_list.append(table)

    # Create dataset
    df = pd.concat(df_list, ignore_index=True)
    df.drop(['52 Week Range'],axis=1,inplace=True)

    return df
```
